```python
# Performing some data manipulations on duration column by converting it into a list

duration = list(df_train["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Checking whether duration contains only of hours or only of minutes
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 to minutes at the end 
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 to hours at the end 

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))     # Extracting the hours for duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))    # Extracting the minutes for duration
```

```python
# Adding duration_hours and duration_mins list to train_data dataframe

df_train["Duration_in_hours"] = duration_hours
df_train["Duration_in_mins"] = duration_mins
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(X_train, y_train)
```

```python
y_pred = reg_rf.predict(X_test)
```

```python
reg_rf.score(X_train, y_train)
```

```python
reg_rf.score(X_test, y_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
